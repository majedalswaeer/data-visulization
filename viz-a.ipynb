## `Importing Libraries`

```python
import pandas as pd
import seaborn as sns
```

## `Selecting data set from seaborn`

```python
sns.get_dataset_names()
my_df=sns.load_dataset('mpg')
my_df.head()
```

## `I imported a dataset showing the miles per gallon for different data types with more features`

```python
# Showing some info about data
my_df.corr()
my_df.describe()
```

```python
d_1=sns.relplot(data=my_df,x='model_year',y="mpg",hue="horsepower")
```

### `from the figure above we can see that in the 80's cars was the highest mpg that car can do, thats because tveryday the tech was improving the fuel consumbtion`

```python
d_2=sns.relplot(data=my_df,x='model_year',y="horsepower",hue="cylinders")
```

### `in the figure above we can notice that the old cars was consuming alot of fuel due to the numbers of cylinder, as a result, these cars also will have more horse powers`

```python
d_3=sns.relplot(data=my_df,x='horsepower',y="origin",hue="cylinders")
```

### I`n the figure above we can notice that the american cars tends to have big engines with high hourse powers than the japanies and eurioian car, in the other hand, the american cars consuming so much fuel than the other cars, and thats what the american cars famous with, THE BIG ENGINES.`

```python
d_4=sns.barplot(data=my_df,x="cylinders", y="weight", palette="Blues_d")
```

### `from the figure above we can notice that the more weight is the car, the more cylinders it has`

```python
d_5=sns.catplot(y="model_year", x="cylinders", data=my_df,hue="origin",jitter=False)
```

### `In the figure above, we can see that the cars from Europe and Japan have cylinders between 4-6, which tends to be more fuel-efficient vehicles; on the consumption side, we will undoubtedly choose the Japanese and European cars`
