```python
import pandas as pd
import seaborn as sns
```

## `Selecting data set from seaborn`

```python
sns.get_dataset_names()
df=sns.load_dataset('titanic')
my_df=df[df['survived']==1]
my_df
```

### `The data presented above has been filtered to include only those who survived the titanic crises.`

```python
my_df.describe()
```

### `From the descripton above we can see that 342 persons has been survived and the older one was 80 years old` 

```python
d_1=sns.countplot(data=my_df,x='class')
```

### `From the figure above we can notice that the most survivers was from the first class, because the lifeboats was easy accses for them.`

```python
d_2=sns.countplot(data=my_df,x='sex')
```

### `we can notice above that the females almost double the males was survived.`

```python
d_3=my_df.groupby(['sex'],as_index=False).mean()
sns.barplot(data=d_3,x='sex',y="age")
females=my_df['sex']=='female'
males=my_df['sex']=='male'
my_df[females].describe()
my_df[males].describe()
```

### `We can notice that almost of the survivers are youngs due to the mean age of females and males`
